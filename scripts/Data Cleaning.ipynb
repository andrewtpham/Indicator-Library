```python
arr[1]
baseline = 'Baseline'
endline = 'Endline'
annual = 'Annual'
quarterly = 'Quarterly'
midline = 'Midline'
monthly = 'Monthly'
bimonthly = 'Bimonthly'
EOC = 'End of cycle'
s_annual = 'Semi-Annual'
baseline_arr = []
endline_arr = []
annual_arr = []
quarterly_arr = []
monthly_arr = []
bimonthly_arr = []
EOC_arr = []
s_annual_arr = []
bc, ac, ec,qc,mc,bmc,eoc,sac = 0,0,0,0,0,0,0,0
```

```python
for i in range(len(arr)):
    bc, ac, ec,qc,mc,bmc,eoc,sac = 0,0,0,0,0,0,0,0
    if pd.isnull(arr[i]):
        arr[i] = 'not available'
    if baseline in arr[i]:
        baseline_arr.append(1)
        bc = 1
    if endline in arr[i]:
        endline_arr.append(1)
        ec = 1
    if annual in arr[i]:
        annual_arr.append(1)
        ac = 1
    if quarterly in arr[i]:
        quarterly_arr.append(1)
        qc = 1
    if monthly in arr[i]:
        monthly_arr.append(1)
        mc = 1
    if bimonthly in arr[i]:
        bimonthly_arr.append(1)
        bmc = 1
    if EOC in arr[i]:
        EOC_arr.append(1)
        eoc = 1
    if s_annual in arr[i]:
        s_annual_arr.append(1)
        sac = 1
    if bc == 0:
        baseline_arr.append(0)
    if ec == 0:
        endline_arr.append(0)
    if ac == 0:
        annual_arr.append(0)
    if bmc == 0:
        bimonthly_arr.append(0)
    if eoc == 0:
        EOC_arr.append(0)
    if mc == 0:
        monthly_arr.append(0)
    if qc == 0:
        quarterly_arr.append(0)
    if sac == 0:
        s_annual_arr.append(0)
```
